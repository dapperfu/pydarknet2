# Minimal ```pydarknet``` detection example.

````python
import sys
import pydarknet2
import numpy as np
````

````python
# Load a classifier.
classifier = pydarknet2.Classifier("cfg/coco.data", "cfg/yolov3.cfg", "weights/yolov3.weights")
classifier
````

````python
image_path = classifier.root+"/data/dog.jpg"
````

````python
import PIL
img = PIL.Image.open(image_path)
img
````

````python
dets = classifier.detect(img)
````
